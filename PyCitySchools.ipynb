```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np


# Load csv file
school_file = "PyCitySchools/Resources/schools_complete.csv"
student_file = "PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_file)
student_df = pd.read_csv(student_file)



# Test read-in
# school_df
# student_df

# Test for col information
# school_df.columns
# student_df.columns





```

```python
#  Merge both tables of data
school_data_complete_df= pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])


# Test merged
# school_data_complete_df
```

```python
# Total schools (begins to look in the merged data for results/this has to be a value_counts for a unique count)
total_schools=school_data_complete_df["school_name"].value_counts()
total_schools.count()
```

```python
# Total students/total number of students
total_students=school_data_complete_df["Student ID"].count()
print (total_students)
```

```python
# Total Budget
total_budget=school_df["budget"].sum()
total_budget
```

```python
# Average Math score
avg_math_score=school_data_complete_df["math_score"].mean()
print (avg_math_score)
```

```python
# Average reading score
avg_read_score=school_data_complete_df["reading_score"].mean()
avg_read_score
```

```python
# % who passed math=score >70
math_pass=school_data_complete_df.loc[(student_df["math_score"] >=70)]
# print (math_pass)

count_pass_math=math_pass["math_score"].count()
# print (count_pass_math)

per_math_class=(count_pass_math/total_students)*100
per_math_class
```

```python
# % students who pass reading=70
read_pass=school_data_complete_df.loc[(student_df["reading_score"] >=70)]
# print (read_pass)

count_pass_read= read_pass["reading_score"].count()
# print (count_pass_read)

per_read_pass=(count_pass_read/total_students)*100
print (per_read_pass)
```

```python
# Overall passing rate=average of the above per_math_pass + per_read_pass / 2
overall_pass=(per_math_class + per_read_pass)/2
print (overall_pass)
```

```python
district_summary = {"Total Schools" : total_schools,
                   "Total Students" : total_students,
                   "Total Budget" : total_budget,
                   "Average Math Score" : avg_math_score,
                   "Average Reading Score" : avg_read_score,
                    "% Passing Math" : per_math_class,
                   "% Passing Reading" : per_read_pass,
                   "% Overall Passing" : overall_pass
                  }


district_summary_df = pd.DataFrame([district_summary])                                              
                                   
district_summary_df = district_summary_df[["Total Schools",
                   "Total Students","Total Budget","Average Math Score", "Average Reading Score",
                   "% Passing Math", "% Passing Reading", "% Overall Passing" ]]

```

```python
copy_school_sum = school_df.rename(columns={"school_name" : "School Name"})

copy_school_sum
```

```python
# delete School ID 
# delete student ID from copy_school_sum
# copy_school_sum = copy_school_sum.drop(columns="Student ID")

del copy_school_sum["School ID"]


copy_school_sum
```

```python
# Calculate new column called "per student budget
copy_school_sum['Per Student Budget']=(copy_school_sum["budget"]/school_df["size"])

# copy_school_sum['Per Student Budget'] = copy_school_sum['Per Student Budget'].map("${:.2f}".format)

copy_school_sum
```

```python
# merge average math tbl with school_df
avg_math_read_tbl = renamed_student_df.groupby(["School Name"])["reading_score", "math_score"].mean().reset_index()

avg_math_read_tbl
```

```python
# merge average math table with copy_school_sum
copy_school_sum = copy_school_sum.merge(avg_math_read_tbl, on="School Name", how="outer")


copy_school_sum
```

```python
# calc total school budget

copy_school_sum["Total School Budget"]= (copy_school_sum["size"]*copy_school_sum["Per Student Budget"])

# copy_school_sum["Total School Budget"] = copy_school_sum["Total School Budget"].map("${:.2f}".format)


copy_school_sum


```

```python
# We will use conditionals to find the % Passing Math and % Passing Reading
# Then, we will store them as new variabels
# We are using a passrate of 70% or higher

# % Passing Reading
summary_passing_read = renamed_student_df[renamed_student_df['reading_score']>=70]

# % Passing Math
summary_passing_math = renamed_student_df[student_df['math_score']>=70]

summary_passing_read
# summary_passing_math
```

```python
# count the number of students passing in reading
pass_read_count_sum = summary_passing_read.groupby(["School Name"])["reading_score"].count().reset_index()

# rename reading_score to Reading Count---different from values in average table above

pass_read_count_sum=pass_read_count_sum.rename(columns={"reading_score":"Reading Count"})



pass_read_count_sum
```

```python
# count the number of students passing in math
pass_math_count_sum = summary_passing_math.groupby(["School Name"])["math_score"].count().reset_index()

# rename reading_score to Reading Count---different from values in average table above

pass_math_count_sum = pass_math_count_sum.rename(columns={"math_score":"Math Count"})



pass_math_count_sum
```

```python
# merge  pass_math_count_sum with pass_read_count_sum
# merge both on School Name and by "inner" to include only contents found in both columns
pass_count = pass_math_count_sum.merge(pass_read_count_sum, on="School Name", how="inner")

pass_count
```

```python
# merge copy_school_sum with pass_count we just updated
# we want to merge on School Name and bu outer to include all
copy_school_sum = copy_school_sum.merge(pass_count, on="School Name", how="inner")

copy_school_sum

```

```python
# % passing reading
# Take the subject count and divide by the school size then * by 100

# % passing math
copy_school_sum["% Passing Math"]= (copy_school_sum["Math Count"]/copy_school_sum["size"])*100


# % passing reading
copy_school_sum["% Passing Reading"]= (copy_school_sum["Reading Count"]/copy_school_sum["size"])*100

copy_school_sum
```

```python
# Delete Math Count and Reading Count
del copy_school_sum["Math Count"]
del copy_school_sum["Reading Count"]


copy_school_sum
```

```python
# Calc % Overall Passing
copy_school_sum["% Overall Passing"] = (copy_school_sum["% Passing Math"] + copy_school_sum["% Passing Reading"])/2

copy_school_sum
```

```python
# rename column for average reading and average math scores
# Overall Passing Rate 

copy_school_sum = copy_school_sum.rename(columns={"reading_score" : "Avg. Reading Score", "math_score" : "Avg Math Score"})


copy_school_sum

```
